```python
columns = ('appId', 'name')
df_games_names = pd.concat([pd.read_csv('../../data/page{}_games.csv'.format(i), usecols=[1, 2], names=columns) for i in range(1, 6)],
ignore_index=True)
```

```python
columns = ('appId', 'user', 'rating', 'hours')
df_reviews = pd.concat([pd.read_csv('../../data/page{}_reviews.csv'.format(i), usecols=[0, 3, 6, 7], names=columns) for i in range(1, 6)], ignore_index=True)
```

```python
df_sentiments = pd.read_csv('../../data/sentiment.csv', names=['appId', 'user','sentiment'])
```

```python
df_reviews['sentiment'] = df_sentiments['sentiment']
```

```python
indexNames = df_reviews[df_reviews['user'] == '#NAME?' ].index
df_reviews.drop(indexNames , inplace=True)
df_unique_reivews = df_reviews.drop_duplicates(subset=['appId', 'user'])
df_unique_reivews.shape
```

```python
g = df_unique_reivews.groupby('user').agg('count')
df_unique_reivews = df_unique_reivews[df_unique_reivews['user'].isin(g[g['rating'] > 1].index)]
```

```python
df_games_users = df_unique_reivews.pivot(
    index='appId',
    columns='user',
    values='sentiment'
).fillna(0)
```

```python
hashmap = {appId: i for i, appId in enumerate(list(df_games_users.index))}
reverse_hashmap = {v: k for k, v in hashmap.items()}
```

```python
'''
    네트워크 시각화를 위한 result_to_links(), write_net_json() 함수 정의
'''
import json

# recommend result를 nodes, links의 dict list로 작성
def result_to_links(appId, raw_recommends, reverse_hashmap):
    # {"source": "src_name", "target": "tgt_name", "dist": dist_value} 형식의 dict list 작성
    for i, (idx, dist) in enumerate(raw_recommends):
        src=df_games_names[df_games_names['appId']==appId].iloc[0].at['name']
        tgt=df_games_names[df_games_names['appId'] == reverse_hashmap[idx]].iloc[0].at['name']
        links.append({"source": src, "target": tgt, "value": dist})

# nodes, lists를 gameNetwork.json 파일로 작성
def write_net_json(nodes, links):    
    # dictionary에 nodes와 links 합치기
    net_dict=dict()
    net_dict['nodes']=nodes
    net_dict['links']=links

    with open('gameNetwork.json', "w") as json_file:
        json.dump(net_dict, json_file)
```

```python
'''
    Connected component labeling을 위해 Graph 클래스 정의
'''
class Graph:

	# init function to declare class variables
	def __init__(self, V):
		self.V = V
		self.adj = [[] for i in range(V)]

	def DFSUtil(self, temp, v, visited):

		# Mark the current vertex as visited
		visited[v] = True

		# Store the vertex to list
		temp.append(v)

		# Repeat for all vertices adjacent
		# to this vertex v
		for i in self.adj[v]:
			if visited[i] == False:

				# Update the list
				temp = self.DFSUtil(temp, i, visited)
		return temp

	# method to add an undirected edge
	def addEdge(self, v, w):
		self.adj[v].append(w)
		self.adj[w].append(v)

	# Method to retrieve connected components
	# in an undirected graph
	def connectedComponents(self):
		visited = []
		cc = []
		for i in range(self.V):
			visited.append(False)
		for v in range(self.V):
			if visited[v] == False:
				temp = []
				cc.append(self.DFSUtil(temp, v, visited))
		return cc

# game 81개 Graph 생성 
g=Graph(81)
```

```python
from scipy.sparse import csr_matrix

sparse_mat_games_users = csr_matrix(df_games_users.values)
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)

model.fit(sparse_mat_games_users)
```

```python
# nn_recommend_games() 함수 정의
#   : sparse_mat_games_users로 fit된 NearestNeighbors모델 fitted_nn과
#     유사게임 예측 대상인 appId를 인자로 받음.
def nn_recommend_games(fitted_nn, appId):
    distances, indices = fitted_nn.kneighbors(
                sparse_mat_games_users[hashmap[appId]],
                n_neighbors=1+1)

    raw_recommends = \
            sorted(
                list(
                    zip(
                        indices.squeeze().tolist(),
                        distances.squeeze().tolist()
                    )
                ),
                key=lambda x: x[1]
            )[1:]

    print('\tRecommendations for {}:'.format(appId))
    for i, (idx, dist) in enumerate(raw_recommends):
        print('{}: {}, distance of {}'.format(i+1, df_games_names[df_games_names['appId'] == reverse_hashmap[idx]].iloc[0].at['name'], dist))
        # print('{}: {}, distance of {}'.format(i+1, reverse_hashmap[idx], dist))
        g.addEdge(hashmap[appId], idx)

    # 추천결과를 links[]에 append
    result_to_links(appId, raw_recommends, reverse_hashmap)    
```

```python
links=[]
for i, appId in enumerate(list(df_games_users.index)):
    nn_recommend_games(model, appId)

cc = g.connectedComponents()
```

```python
for i, (li) in enumerate(cc):
    print(f'{i}: {li}')
```

```python
# {"id": "game_name", "group": group_num} 형식의 dict list 작성
nodes=[]
for i, (li) in enumerate(cc):
    for idx in li:
        name=df_games_names[df_games_names['appId'] == reverse_hashmap[idx]].iloc[0].at['name']
        nodes.append({"id":name, "group":i})
```

```python
# json 파일 작성
write_net_json(nodes, links)
```

```python
'''
    minmax scaling을 위한 작업들
'''
dist_values=[]
for dic in links:
    dist_values.append(dic['value'])
```

```python
from sklearn.preprocessing import minmax_scale

# dist_value list를 np array로 변환
dist_arr=np.asarray(dist_values, dtype=np.float32)
# minmax scaling
dist_MinMax_scaled = minmax_scale(dist_arr, axis=0, copy=True)
# 다시 list로..?
dist_MinMax_scaled=dist_MinMax_scaled.tolist()
```

```python
### links의 value를 minmax scaling된 값으로 수정
for i, (dic) in enumerate(links):
    dic['value']=dist_MinMax_scaled[i]
```

```python
# json 파일 다시 작성
write_net_json(nodes, links)
```
